# Monte Carlo Simulation of the probability of drawing spades
You are to draw a card 13 times from a standard 52-card deck, replacing it after each draw. How many times do you draw a spade?

What I want the code to do:
- Receive an argument : number of iterations (of 13 draws)
- Generate a distribution table containing cumulative frequency of spades, running average and optionally display that table
- Plot a scatterplot of draws against trial
- Plot a cummulative average number of draws for the entire simulation

```
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('Imports complete')
```

```
def monte_carlo_spades(n, show_df=False):
  '''
  Print a scatter plot of the number of spades and an ogive of the running_average of spades turned up
  Return a pandas dataframe

  Parameters
  ----------
  n : Integer
      The number of times to run the simulation
  show_df : Boolean
      Choose whether to print the resulting dataframe
  '''
  # create arrays for trial, number of spades in trial and cumulative number of spades
  trial = np.arange(1, n+1)
  spades = np.random.randint(low=0, high=14, size=n)
  cumulative = np.cumsum(spades)
  
  #create pandas dataframe
  df = pd.DataFrame({'trial':trial,
                     'spades':spades,
                     'cumulative':cumulative})
  df['running_average'] = np.round(df['cumulative']/df['trial'], decimals=2)
  if show_df: print(df)

  plt.style.use('ggplot')

  #plot scatter
  sns.scatterplot(x=trial, y=spades, data=df)
  plt.xlabel('Trial')
  plt.ylabel('Number of spades')
  plt.yticks(ticks=range(0,14))
  plt.title('Number of spades drawn for every trial')
  plt.show()

  #plot ogive
  plt.clf()
  sns.lineplot(x=trial, y='running_average', data=df)
  plt.ylabel('Cumulative average number of spades')
  plt.xlabel('Trial')
  plt.title('Average number of spades throughout the simulation')
```

```
monte_carlo_spades(40, show_df=True)
```
